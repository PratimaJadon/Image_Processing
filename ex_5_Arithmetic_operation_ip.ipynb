# Set threshold and maximum value

```python
# Set threshold and maximum value
thresh = 150
maxValue = 255
```

```python
# Binary threshold
th, dst = cv2.threshold(img, thresh, maxValue, cv2.THRESH_BINARY)
```

```python
plt.imshow(dst, cmap='gray')
plt.show()
```

# **Bitwise operations**
are used in image manipulation and used for extracting essential parts in the image. In this article, Bitwise operations used are :


AND,
OR,
XOR,
NOT

```python
img_and = cv2.bitwise_and(img, img1, mask = None)
```

```python
# AND operation
from google.colab.patches import cv2_imshow
cv2_imshow(img_and)
```

```python
#Bitwise OR operation

cv2.bitwise_or(img,img1,  mask=None)
```

```python
#Bitwise XOR operation
cv2.bitwise_xor(img,img1 , mask= None)
```

```python
#Bitwise NOT operation

cv2.bitwise_not(img,img1, mask=None)
```
